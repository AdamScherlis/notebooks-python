```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

# Fetch data
sp500 = yf.download('^GSPC', start='2020-01-01')
vix = yf.download('^VIX', start='2020-01-01')

# Create plot
plt.figure(figsize=(12, 6))
plt.plot(sp500.index, sp500['Close'], label='S&P 500')
plt.plot(vix.index, vix['Close'], label='VIX')

# Customize plot
plt.title('S&P 500 and VIX Historical Data')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Show plot
plt.show()
```

```python
# Calculate daily returns
sp500['Returns'] = sp500['Close'].pct_change() * 100

# Compute 30-day rolling standard deviation of returns
sp500['Volatility'] = sp500['Returns'].shift(-30).rolling(window=30).std()

# Annualize the volatility (assuming 252 trading days in a year)
sp500['Annualized_Volatility'] = sp500['Volatility'] * np.sqrt(252)

# Plot the results

plt.figure(figsize=(12, 6))
plt.plot(sp500.index, sp500['Annualized_Volatility'], label="SP500 actual")
plt.plot(vix.index, vix['Close'], label='VIX')
plt.title('S&P 500 30-Day Rolling Volatility')
plt.ylim(0, None)
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.show()


plt.figure(figsize=(12, 6))
plt.scatter(vix['Close'], sp500['Annualized_Volatility'], label="SP500")
#plt.plot(vix.index, , label='VIX')
plt.title('S&P 500 30-Day Rolling Volatility')
plt.ylim(0, None)
plt.xlim(0, None)
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.show()


plt.figure(figsize=(12, 6))
plt.plot(sp500.index, sp500['Volatility'], label="Vol")
plt.plot(sp500.index, sp500['Returns'], label='Returns')
plt.title('S&P 500 30-Day Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.show()
```

```python
from datetime import *
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = [
    (datetime(2020, 11, 18),  2),
    (datetime(2020, 12,  3), 22),
    (datetime(2020, 12, 28), 40),
    (datetime(2021,  1,  5), 47),]
data2 = [
    (datetime(2020, 11, 26), 189),
    (datetime(2021,  1, 14), 190),
    (datetime(2021,  3, 25), 200)
]
data3 = [data[0], data2[-1]]
```

```python
plt.plot([d[0] for d in data], [d[1] for d in data])
plt.plot([d[0] for d in data2], [d[1] for d in data2])
plt.plot([d[0] for d in data3], [d[1] for d in data3])
```

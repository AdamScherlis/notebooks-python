```python
s = '1'

for _ in range(128):
    print(s[:64]*(64//len(s[:64])))
    # hide 1 prefix
    l = [len(t)+1 for t in s.split('1')[1:]]
    
    #print(l)
    
    if len(l) == 1:
        l += l
    assert len(l) % 2 == 0
    
    # hidden 1 prefix produces hidden 0 in bitstring
    bit = 1
    s = ''
    for n in l:
        s += '01'[bit]*n
        bit = 1-bit
```

```python
l = [1]

for _ in range(64):
    #print(l)
    
    # implicit tail of 0
    s = ''.join(['0'*(n-1)+'1' for n in l])
    
    print(s)
    
    # decode RLE
    zs = '0'+s
    l = []
    b = '0'
    a = 0
    while zs:
        if zs[0] != b:
            b = '10'[int(b)]
            l += [a]
            a = 0
        a += 1
        zs = zs[1:]
    l += [a]
```

```python
def zero_path(h, v, r=100):
    return ""
def one_path(h, v, r=100):
    return f"l 0 {v*r} l {h*r} 0"
def two_path(h, v, r=100):
    return f"q {h*(r*4)//5} 0 {h*r} {v*r}"
def three_path(h, v, r=100):
    return f"l 0 {v*r} h {h*r//4} a {(r*3)//4} {r//2} 0 {(1-v)//2} {(1-h*v)//2} 0 {-v*r} h {-h*r//4} z"
```

```python
def path_spec(k, h, v, r=100):
    return ["",
           f"l 0 {v*r} l {h*r} 0",
           f"q {h*(r*4)//5} 0 {h*r} {v*r}",
           f"l 0 {v*r} h {h*r//4} a {(r*3)//4} {r//2} 0 {(1-v)//2} {(1-h*v)//2} 0 {-v*r} h {-h*r//4} z"][k]
```

```python
def lone_glyph(n, r=20):
    ks = [n%4, [0, 2, 1, 3][(n>>2)%4], (n>>4)%4, [0, 2, 1, 3][(n>>6)%4]]
    c = int(1.1*r)
    out = f"""<svg width="{2*c}" height="{2*c}">
<path d="M {c} {c-r} l 0 {2*r}" \
stroke="black" stroke-width="{r//5}" fill="none" />"""
    for i in range(4):
        out += f"""<path d="M {c} {c} {path_spec(ks[i], (-1)**(i//2), (-1)**i, r)}" \
stroke="black" stroke-width="{r//5}" fill="none" />"""
    out += """</svg>"""
    return out
```

```python
def glyph_row(ns, r=20):
    c = int(1.1*r)
    out = f"""<svg width="{2.5*c*len(ns)}" height="{2*c+5}">"""
    for x, n in enumerate(ns):
        ks = [n%4, [0, 2, 1, 3][(n>>2)%4], (n>>4)%4, [0, 2, 1, 3][(n>>6)%4]]
        out += f"""<path d="M {(2.5*x+1)*c} {c-r} l 0 {2*r}" \
    stroke="black" stroke-width="{r//5}" fill="none" />"""
        for i in range(4):
            out += f"""<path d="M {(2.5*x+1)*c} {c} {path_spec(ks[i], (-1)**(i//2), (-1)**i, r)}" \
    stroke="black" stroke-width="{r//5}" fill="none" />"""
    out += """</svg>"""
    return out
```

```python
# self-describing string
# 16 hex values per line
# inspired by a bug or something
# mid-2021
```

```python
def ttf(s):
    byts = [ord(c) for c in s]
    rows = range((len(s)-1)//16+1)
    return '\n'.join(
        [''.join(
            [(' ' * (b < 100)) + str(b) + ',' for b in byts[row * 16: (row+1) * 16]]
        ) for row in rows]
    )
```

```python
s = ' '
for i in range(10):
    print(s)
    print()
    s = ttf(s)
```

```python
# 2022-07
# comparing Roman numerals to an "optimized" version
# (re-skinned octal)
# underscore is a left-combining overbar
```

```python
def roman(n):
    out = ''
    if n == 0:
        return out
    digs = enumerate(reversed(str(n)))
    parts = ['IVX', 'XLC', 'CDM']
    for k, d in digs:
        i, v, x = parts[k%3]
        i += '_' * (k//3)
        v += '_' * (k//3)
        x += '_' * (k//3)
        if k%3 == 0 and k > 0 and int(d) <= 3:
            i = 'M' + '_' * (k//3 - 1)
        out = ['', i, i+i, i+i+i, i+v, v, v+i, v+i+i, v+i+i+i, i+x][int(d)] + out
    return out
```

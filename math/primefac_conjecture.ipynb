```python
from sympy.ntheory import factorint
```

```python
factorint(12)
```

```python
def mutate(n):
    facs = factorint(n)
    
    x = 0
    for p in facs:
        x += p * facs[p]
        
    if x == n and n in facs:
        return x + 2
    return x
```

```python
for n in range(5, 1_000_000):
    its = 0
    ns = [n]
    while len(ns) < 30 and n != 5:
        n = mutate(n)
        ns.append(n)
        its += 1
    if len(ns) > 15:
        print(len(ns), ns)
```

```python
def mutwo(n):
    facs = factorint(n)
    
    x = 0
    for p in facs:
        x += p * facs[p]
        
    if x == n and n in facs:
        return x + 1
    return x
```

```python
for n in range(2, 1000):
    its = 0
    ns = [n]
    while len(ns) < 20 and n != 5:
        n = mutwo(n)
        ns.append(n)
        its += 1
    if len(ns) > 1:
        print(len(ns), ns)
```

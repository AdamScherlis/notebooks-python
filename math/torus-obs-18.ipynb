```python
import networkx as nx
import matplotlib.pyplot as plt

def create_graph_from_upper_triangle(adj_string):
    n = int((1 + (1 + 8*len(adj_string))**0.5) / 2)
    G = nx.Graph()
    G.add_nodes_from(range(n))
    
    idx = 0
    for i in range(n-1):
        for j in range(i+1, n):
            if adj_string[idx] == '1':
                G.add_edge(i, j)
            idx += 1
    
    return G

def draw_optimized_graph(G):
    pos = nx.kamada_kawai_layout(G)
    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=False, node_color='blue', 
            node_size=100, font_size=16, font_weight='bold')
    # edge_labels = {(u, v): f'{u}-{v}' for u, v in G.edges()}
    # nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title("Optimized Graph Layout")
    plt.axis('equal')
    plt.savefig(f'graph{len(G)}C.png')

# Example usage
adj_string = "101101"
G = create_graph_from_upper_triangle(adj_string)
draw_optimized_graph(G)
```

```python

G = create_graph_from_upper_triangle(uppertri)
draw_optimized_graph(G)
```

```python
s = '100000011100000001110100001101000011011001011001011101110111000111'
G = create_graph_from_upper_triangle(s)
draw_optimized_graph(G)
```

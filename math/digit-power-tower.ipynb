# Digit power-tower map

```python
import sys
import numpy as np
```

```python
sys.set_int_max_str_digits(10_000)
```

```python
def trim(n):
    if n < 10:
        return str(n)
    s = str(n).split('1')[0]
    if '00' in s:
        s = s.split('00')[0]
    if '0' in s:
        s = s.split('0')[0][:-1]
    if not s:
        return '1'
    return s

def powertower(n):
    s = trim(n)
    x = 1
    for i, d in enumerate(reversed(s)):
        if x * len(d) > 10_000:
            print(f"too big: {d}**{x} in {s} at {s[-1-i:]}")
            return 0
        x = int(d)**x
    return x

def iterate(n):
    print(n)
    while n > 9:
        n = powertower(n)
        print(f'{trim(n)}{'...' if str(n) != trim(n) else '.'}', flush=True)
    print('done')
```

```python
iterate(28)
```

```python
iterate(742)
```

```python
iterate(543)
```

```python
for n in range(20, 1000):
    print('\n')
    iterate(n)
```

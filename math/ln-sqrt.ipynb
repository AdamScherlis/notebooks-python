```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
def idcolor(x, y, RLOGMAX=1):
    r = np.sqrt(x**2 + y**2)
    t = np.arctan2(y, x)
    return [0.5 + r/RLOGMAX * 0.5 * np.cos(t + k*2*np.pi/3) for k in range(3)]
```

```python
def logcolor(x, y, RLOGMAX=1):
    r = np.sqrt(x**2 + y**2)
    t = np.arctan2(y, x)
    relog = np.log(r)
    imlog = t
    rlog = np.sqrt(relog**2 + imlog**2)
    tlog = np.arctan2(imlog, relog)
    return [0.5 + rlog/RLOGMAX * 0.5 * np.cos(tlog + k*2*np.pi/3) for k in range(3)]
```

```python
x = np.linspace(-2, 2, 100).reshape([1, -1])
y = np.linspace(-2, 2, 100).reshape([-1, 1])

plt.imshow(np.array(logcolor(x, y)).transpose(1,2,0))
```

```python
x = np.linspace(-2, 2, 100).reshape([1, -1])
y = np.linspace(-2, 2, 100).reshape([-1, 1])

plt.imshow(np.array(idcolor(x, y)).transpose(1,2,0))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gk_quant(p):
    return np.floor(2/(2-2**p)-2)+1
```

```python
ps = np.linspace(0,.99999,100000)
plt.plot(np.log(1-ps), np.log(gk_quant(ps)))
```

```python
coefs = gk_quant(np.random.random(10000))
```

```python
def gk_pdf(n):
    return -np.log(1-1/(n+1)**2)/np.log(2)
```

```python
sum(map(lambda p: -p*np.log(p), gk_pdf(np.arange(1,1000))))
```

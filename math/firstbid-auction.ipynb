```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(k, xs):
    # (1-x) * (x + 1/(k*2pi) * (1-cos(k*2pi*x)))
    return (1-xs) * (xs + 1/(k * 2*np.pi) * (1-np.cos(k * 2*np.pi * xs)))
```

```python
xs = np.linspace(0,1,1000)
yss = np.array([fun(k, xs) for k in range(1,6)]).T
```

```python
plt.plot(xs, yss)
```
